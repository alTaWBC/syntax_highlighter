```python
import pandas as pd
array = [
    {'tit': 'polarity', 'clas': ['bom', 'mau'], 'testes': [['guerra', 'mau'], [
        'pizza', 'bom'], ['descançar', 'bom'], ['cansado', 'mau']]},
    {'tit': 'class-animais', 'clas': ['ave', 'mamífero', 'peixe'], 'testes': [['melro', 'ave'], ['águia', 'ave'], ['pomba', 'ave'], ['gato', 'mamífero'], ['leão', 'mamífero'], ['macaco', 'mamífero'], ['boi', 'mamífero'], ['tubarão', 'peixe'], ['sardinha', 'peixe'], ['salmão', 'peixe']]}]

words = ['bom', 'mau', 'ave', 'mamífero', 'peixe'] # placeholder
example_missing_words = ['bom', 'ave', 'mamífero', 'peixe']  # placeholder
dataframe = pd.DataFrame(array)
dataframe
```

```python
# Utils
# Não testei com as funções do modelo mas presumo que de logo

def _remove_non_existing_words(row):
    return all([clas in words for clas in row['clas']])
    # return all([clas in model.wv.key_to_index for clas in classes]) Usar isto

import random # Remover quando remover o return random
random.seed(42)  # Remover quando remover o return random

def _similarity(row):
    # return model.wv.similarity(row['word'], row['testes']) Usar isto
    return random.random()
```

```python
# Unknown_words
all_words = dataframe["clas"].explode()
unknown_words = all_words[~all_words.isin(example_missing_words)] # O ~ nega o dataframe (~[True, False, True] -> [False, True, False])
unknown_words
```

```python
# Remove rows with words not present in the dictionary
dataframe[dataframe.apply(_remove_non_existing_words, axis=1)]
dataframe
```

```python
# Dividir Testes em várias linhas
dataframe = dataframe.explode('testes')  
dataframe
```

```python
# Dividir testes em duas colunas
dataframe[['word', 'answer']] = pd.DataFrame(
    dataframe['testes'].tolist(), index=dataframe.index)
dataframe
```

```python
# Dividir classes em várias linhas
dataframe = dataframe.explode('clas')
dataframe[:10]
```

```python
# predict do modelo
dataframe['predicted'] = dataframe.apply(_similarity, axis=1)
dataframe[:10]
```

```python
# Buscar maior predicted
result = dataframe.groupby(by=['word']).max()
result
```
